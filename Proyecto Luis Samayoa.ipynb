## Funcion de entrenamiento

```python
unos=np.ones_like(x1)


def entreno_funcion(x,y,epochs,imprimir_error_cada,lr):
      
    matriz_base=np.vstack([x,unos])
    iteracion=np.zeros(epochs)
    error=np.zeros(epochs)
    parametros=np.array([0.,0.])
    resultados_simulacion = {}
    
    for i in range(epochs):
        y_hat=np.matmul(parametros,matriz_base)
        iteracion[i]=i
        error[i]=0.5*(np.mean((y-y_hat)**2))
        
        if (i%imprimir_error_cada==0):
            print(error[i])
        
        gradiente=np.array([np.mean((y_hat-y)*x),np.mean((y-y_hat))])
        parametros-=lr*gradiente
        resultados_simulacion[i]=parametros.copy()
    
            
    return error,resultados_simulacion,iteracion
```

## Funcion de grafica de error

```python
def grafica_error(error):
      
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1) 

    ax.grid()
    ax.plot(iteracion,error)

    ax.set_xlabel('iteracion')
    ax.set_ylabel('error')
    ax.set_title('Grafica de error')

    plt.show()
```

```python
def evolucion_modelo(x,y,epochs,imprimir_error_cada,lr):
    
    fig = plt.figure(figsize=(14, 12), dpi= 100)
    plt.scatter(x, y)
    plt.title("Rectas vs tiempo")
    plt.ylabel("Precio de las viviendas")
    plt.xlabel("Iteracion")
    
    
    matriz_base=np.vstack([x,unos])
    iteracion=np.zeros(epochs)
    error=np.zeros(epochs)
    parametros=np.array([0.,0.])

       
    for i in range(epochs):
        y_hat=np.matmul(parametros,matriz_base)    
        error[i]=0.5*(np.mean((y-y_hat)**2))
        gradiente=np.array([np.mean((y_hat-y)*x),np.mean((y-y_hat))])
        parametros-=lr*gradiente

    
        if (i % imprimir_error_cada==0): 
            plt.plot(x,y_hat)
            plt.annotate("n = %d" % (i), (np.max(x),y_hat[i]))      
                
      
```
